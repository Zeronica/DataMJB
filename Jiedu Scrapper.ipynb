```python
import pandas as pd
import numpy as np
import bs4 as bs
import requests


##run the below lines to install google translator package
##import sys
##!{sys.executable} -m pip install googletrans
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.146 Safari/537.36',
    'From': 'avyay1997@gmail.com' 
}

##from googletrans import Translator ## importing translator package
##translator=Translator()

chars=pd.DataFrame()

index_start=0
index_end=1

read=pd.read_csv('data_populated_all_fixed_8000.csv')
array=np.array(read['link'])

while index_start<=index_end:
    good_chars=list()
    bad_chars=list()
    community_link=list()
    website_link=list()
    p=1
    while True:
        source=requests.get(array[index_start]+'/jiedu/p{}'.format(p), headers=headers).content
        soup=bs.BeautifulSoup(source, features='html.parser')
        good_present=soup.find_all(class_='info-character good-character')
        bad_present=soup.find_all(class_='info-character bad-character')
        a=0
        b=0
        if(good_present): ##checkin if given class present or not
            for x in good_present:
                good_chars.append(x.text)
                website_link.append(array[index_start])
        else:
            a=1
            
        if(bad_present):
            for y in bad_present:
                bad_chars.append(y.text)
                
        else:
            b=1
        
        if(a==1 and b==1): ##jeidu class not present
            print('Done with: ',array[index_start])
            break; #terminating nested while loop
            
        data=pd.DataFrame(data=website_link, columns=['Link'])
        data['Good Characters']=good_chars
        data['Bad Characters']=bad_chars
        chars=chars.append(data, ignore_index=True)
        p=p+1
        
    index_start=index_start+1
    

chars.to_csv('Jiedu.csv')    
chars.head()
```
