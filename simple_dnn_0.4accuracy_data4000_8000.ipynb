```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import tensorflow as tf
from sklearn.model_selection import train_test_split
import keras
```

```python
x = pd.read_csv('x.csv',header=None).values[:,1:].astype('float32')
y = pd.read_csv('y.csv',header=None).values[:,1].astype('float32')
y = y.reshape(-1,1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
x = preprocessing.scale(x_train)  
y = preprocessing.scale(y_train)

x_scaler = preprocessing.StandardScaler().fit(x_train)  
y_scaler = preprocessing.StandardScaler().fit(y_train) 
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
model = Sequential()
model.add(Dense(30, activation='relu', input_shape=[x.shape[1]]))
model.add(Dropout(0.1))

model.add(Dense(20, activation='relu'))
model.add(Dropout(0.1))

model.add(Dense(10, activation='relu'))
model.add(Dropout(0.1))

model.add(Dense(5, activation='relu'))
model.add(Dropout(0.1))

model.add(Dense(1))
model.summary()
```

```python
model.compile(loss="mse",
              optimizer='rmsprop',
              metrics=['mse'])
```

```python
model.fit(x, y,
          epochs=3000)
```

```python
import math

total = len(y_test)

cnt = 0  
x_test = preprocessing.scale(x_test)  
predicted = model.predict(x_test)  
for i,v in enumerate(predicted):  
    p = y_scaler.inverse_transform(v)
    r = y_test[i]
    if math.fabs(r - p)/r < 0.15:
        cnt += 1
print("accuracy of test: %d/%d " % (cnt, total), cnt/total)

total = len(y_train)

cnt = 0  
predicted = model.predict(x)  
for i,v in enumerate(predicted):  
    p = y_scaler.inverse_transform(v)
    r = y_train[i]
    if math.fabs(r - p)/r < 0.15:
        cnt += 1
print("accuracy of train: %d/%d " % (cnt, total), cnt/total)
```
